```python
# import packages
import numpy as np
from keras import models, layers
import matplotlib.pyplot as plt
```

## Prepare data

### Load data

```python
from keras.datasets import cifar10

(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

### Explore data

```python
print("x_train shape: %s" % (str(x_train.shape)))
print("y_train shape: %s" % (str(y_train.shape)))
print("x_test shape: %s" % (str(x_test.shape)))
print("y_test shape: %s" % (str(y_test.shape)))
```

```python
images_indices = []
# get the first 5 index of every class label 
for i in range(10):
  for j in range(5):
    images_indices.append(np.argwhere(y_train == i)[j][0].item())
    
labels = ['airplane',	'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']										
# plot example of every class label
rows = 10
f = plt.figure(figsize=(30,30))
for i in range(len(images_indices)):
    sp = f.add_subplot(rows, len(images_indices)//rows, i+1)
    sp.axis('Off')
    sp.set_title(labels[i//5], fontsize=16)
    plt.imshow(x_train[images_indices[i]], cmap=plt.cm.binary)        
```

### Preprocess data

```python
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255

# categorically encode the labels
from keras.utils import to_categorical

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

## Build models

### Helper functions

```python
def train(model):
  """
  Compile and train the model
  """
  model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
  return model.fit(x_train, y_train, epochs=15, batch_size=64, validation_data=(x_test,y_test))
```

```python
def plot(hist):
  """
  Plot the training/validation loss/accuracy graphs  
  """
  loss = hist.history['loss']
  val_loss = hist.history['val_loss']
  epochs = range(1, len(loss) + 1)

  plt.figure()
  plt.plot(epochs, loss, 'bo', label='Training loss')
  plt.plot(epochs, val_loss, 'b', label='Validation loss')
  plt.xlabel('Epochs')
  plt.ylabel('Loss')
  plt.title('Training and validation loss')
  plt.legend()
  plt.show()

  acc = hist.history['acc']
  val_acc = hist.history['val_acc']

  plt.plot(epochs, acc, 'bo', label='Training acc')
  plt.plot(epochs, val_acc, 'b', label='Validation acc')
  plt.title('Training and validation accuracy')
  plt.xlabel('Epochs')
  plt.ylabel('Acc')
  plt.legend()
  plt.show()
```

### Convnet inspired by AlexNet

```python
model = models.Sequential()
# convolution layer & pooling layer
model.add(layers.Conv2D(64, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(96, (3, 3), activation='relu'))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.Conv2D(96, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Flatten())
# # 2 dense layer
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(512, activation='relu'))

# # softmax
model.add(layers.Dense(10, activation='softmax'))
model.summary()
```

```python
hist = train(model)
plot(hist)
```

With this convnet, we get over 70% accuracy on our validation set. The improvement of the training acc is slow, only reaches above 80% after 15 epochs. This model also overfits. With the fashion MNIST, a similar model will get over 90% accuracy. The images are not grayscale so it's harder to train on this CIFAR10 dataset. 

### Fight overfitting

#### Add dropout layers

We add dropout layers right after the dense layers.

```python
model = models.Sequential()
# convolution layer & pooling layer
model.add(layers.Conv2D(64, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(96, (3, 3), activation='relu'))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.Conv2D(96, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Flatten())
# # 2 dense layer
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dropout(0.3))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dropout(0.4))

# # softmax
model.add(layers.Dense(10, activation='softmax'))
model.summary()
```

```python
hist = train(model)
plot(hist)
```

With dropout layers, the model even struggles to get the training accuracy pass 70% accuracy. 

#### Add batch normalization

```python
model = models.Sequential()
# convolution layer & pooling layer
model.add(layers.Conv2D(64, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(96, (3, 3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Conv2D(96, (3, 3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Flatten())
# # 2 dense layer
model.add(layers.Dense(512, activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Dropout(0.3))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Dropout(0.4))

# # softmax
model.add(layers.Dense(10, activation='softmax'))
model.summary()
```

```python
hist = train(model)
plot(hist)
```

The training accuracy improved very fast compared to previous models. The validation accuracy also reached 75%. I'm happy with this result because I also struggled to guess the images.
